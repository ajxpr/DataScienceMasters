**Question 4**: What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

In Pandas, a DataFrame is a two-dimensional labeled data structure that can store data of different types (integer, float, string, etc.). It is similar to a table or a spreadsheet where data is organized in rows and columns. Each column in a DataFrame is represented by a Pandas Series, and together they form the DataFrame.

A Pandas Series, on the other hand, is a one-dimensional labeled array that can hold data of a single data type. It can be thought of as a single column of a DataFrame.

```python
# Example

import pandas as pd

# Create a DataFrame
data = {
    'Name': ['John', 'Emily', 'David'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Sydney']
}
df = pd.DataFrame(data)

# Accessing a column as a Series
name_series = df['Name']
print(name_series)
```

**Question 5**: What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

There are several common functions in Pandas that can be used to manipulate data in a DataFrame. Some of them include:

**head()**: This function returns the first n rows of a DataFrame. It is useful to quickly preview the data or get a glimpse of the structure.

```python
# Example

df.head(5)
```

**tail()**: This function returns the last n rows of a DataFrame. It can be used to check the end of the data or verify the data structure.

```python
# Example

df.tail(3)
```

**describe()**: This function provides statistical summary of the DataFrame, including count, mean, min, max, etc. It gives a quick overview of the data distribution.

```python
# Example

print(df.describe())
```

**info()**: This function provides concise information about the DataFrame, including column names, data types, and memory usage. It helps in understanding the data types and missing values.

```python
# Example

df.info()
```

**sort_values()**: This function sorts the DataFrame by one or more columns. It is useful for ordering the data based on specific criteria.

```python
# Example

import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [25, 32, 18, 45, 28],
        'Salary': [50000, 70000, 30000, 90000, 60000]}

df = pd.DataFrame(data)

# Sort the DataFrame by the 'Age' column in ascending order
df_sorted = df.sort_values('Age')

print(df_sorted)
```

**Question 6**: Which of the following is mutable in nature Series, DataFrame, Panel?

Among the options given, both Series and DataFrame are **mutable** in nature, while Panel is **not mutable**.

- **Series**: A Series in Pandas is a one-dimensional labeled array that can hold any data type. It is mutable, which means its values can be modified after creation.

- **DataFrame**: A DataFrame in Pandas is a two-dimensional labeled data structure with columns of potentially different data types. It is also mutable, allowing modifications to its values, columns, and structure.

- **Panel**: A Panel is a three-dimensional data structure in Pandas, but it is considered deprecated and no longer recommended for general use. Panels are not mutable. The preferred alternative for handling three-dimensional data is to use MultiIndex DataFrame structures.

**Question 7**: Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd

# Create two Series
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series(['a', 'b', 'c', 'd', 'e'])

# Create a DataFrame using the Series
df = pd.DataFrame({'Column1': s1, 'Column2': s2})

# Print the DataFrame
print(df)
```
