**Question 1**: What is Flask Framework? What are the advantages of Flask Framework?

Flask is a micro web framework for Python. It is designed to be lightweight, flexible, and easy to use, making it a popular choice for building web applications and APIs. 

Here are some advantages of Flask:

1. **Lightweight and Minimalistic**: Flask follows a minimalist design philosophy, providing only the essential features needed for web development. It has a small codebase and a simple core, which makes it easy to understand and use.

2. **Easy to Learn**: Flask has a straightforward and intuitive API that makes it easy for developers, especially beginners, to get started quickly. Its simplicity allows for rapid development and reduces the learning curve.

3. **Flexibility and Customization**: Flask allows developers to have granular control over the components and features of their application. It provides a modular structure, allowing you to choose and integrate only the necessary extensions and libraries as per your project requirements.

4. **Extensible and Scalable**: Flask provides a wide range of extensions and libraries that can be easily integrated to add functionality to your application. It supports the use of third-party libraries and can be extended to handle complex tasks and scale with the growth of your project.

5. **Built-in Development Server**: Flask comes with a built-in development server, which makes it convenient for testing and debugging your application during development. It allows you to quickly iterate and see the changes in real-time.

**Question 3**: What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (routes) to specific functions in your application. It allows you to define different URL patterns and associate them with corresponding functions that handle the request and generate a response.

The @app.route() decorator in Flask is used to define app routes. It is placed above a function definition and specifies the URL pattern that should trigger that function. For example, @app.route('/') associates the root URL (/) with a particular function.

**Question 4**: Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:

**Company Name**: ABC Corporation

**Location**: India

**Contact Detail**: 999-999-9999

Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

@app.route('/')
def company_details():
    return '''
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999
    '''

if __name__ == '__main__':
    app.run()
```

**Question 5**: What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It is a helper function that generates a URL for the specified endpoint.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "Home Page"

@app.route('/about')
def about():
    return "About Page"

@app.route('/contact')
def contact():
    return "Contact Page"

if __name__ == '__main__':
    with app.test_request_context():
        # Using url_for to generate URLs
        home_url = url_for('home')
        about_url = url_for('about')
        contact_url = url_for('contact')

        # Printing the generated URLs
        print("Home URL:", home_url)
        print("About URL:", about_url)
        print("Contact URL:", contact_url)
```
