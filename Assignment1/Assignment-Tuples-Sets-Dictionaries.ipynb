**Question 2**: What are the two tuple methods in python? Give an example of each method. Give a reason why
tuples have only two in-built methods as compared to Lists.

Python has only two tuple methods: `index()` and `count()`.

Examples of each method:

```python
# index()
my_tuple = ('apple', 'banana', 'orange', 'banana')
index = my_tuple.index('banana')
print(index)
```

```python
# count()
my_tuple = (1, 2, 3, 4, 2, 2)
count = my_tuple.count(2)
print(count)
```

Tuples have fewer built-in methods compared to lists due to their immutability.

**Question 4**: Explain the difference between the union() and update() methods for a set. Give an example of each method.

`update()` modifies the original set in place by adding elements from another set or iterable. On the other hand `union()` creates a new set.

Example of each method:

```python
# update()
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.update(set2)
print(set1)
```

```python
# union()
set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = {5, 6, 7}

union_set = set1.union(set2, set3)
print(union_set)
```

**Question 8**: What are the three view objects in dictionaries? Use the three in-built methods in python to display these three view objects for the given dictionary.

dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}

The three view obkects in dictionaries are:
1. keys()
2. values()
3. items()

```python
#keys()
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
keys_view = dict1.keys()
print(keys_view)
```

```python
# values()
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
values_view = dict1.values()
print(values_view)
```

```python
# items()
dict1 = {'Sport': 'Cricket', 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
items_view = dict1.items()
print(items_view)
```
